### <span style="font-family: 'D2Coding'">1.6.1 단순한 그래프 그리기</span>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 데이터 준비
x = np.arange(0, 6, .1)    # 0에서 6까지 0.1 간격으로 생성
y = np.sin(x)
```

```python
# 그래프 그리기
plt.plot(x, y)
plt.show()
```

### <span style="font-family: 'D2Coding'">1.6.2 pyplot의 기능</span>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 데이터 준비
x = np.arange(0, 6, .1)    # 0에서 6까지 0.1 간격으로 생성
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# 그래프 그리기
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title('sin & cos')
plt.legend()
plt.show()
```
